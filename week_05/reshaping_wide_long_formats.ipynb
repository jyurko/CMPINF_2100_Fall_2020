## Reshape with a summary statistics

Long to wide, we just reorganized. But we could GROUP and SUMMARIZE when we reshape. This is done by `.pivot_table()`.

```python
air_quality
```

```python
air_quality.pivot_table(values='value',
                       index='location',
                       columns='parameter',
                       aggfunc='mean')
```

```python
air_quality.pivot_table(values='value',
                       index='location',
                       columns='parameter',
                       aggfunc='mean',
                       margins=True)
```

We could have done this with `.groupby()`.

```python
air_quality.groupby(['parameter', 'location']).mean()
```

```python
air_quality.groupby(['parameter', 'location']).mean().\
reset_index().\
set_index('location').\
pivot(columns='parameter', values='value')
```
