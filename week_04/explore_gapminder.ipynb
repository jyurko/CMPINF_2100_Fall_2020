### What are the unique values?

```python
### that's where the .value_counts()

gap_df['continent'].value_counts()
```

```python
### visualize the number of rows/entries associated with each unique value

fig, ax = plt.subplots()

gap_df['continent'].value_counts().plot(kind='bar', ax=ax)

plt.show()
```

```python
### horizontal bar chart
fig, ax = plt.subplots(figsize=(12, 6))

gap_df['continent'].value_counts().plot(kind='barh', ax=ax)

plt.show()
```

```python
### print out the unique values for country
gap_df['country'].value_counts()
```

```python
### what's going on with the `year`
gap_df['year'].value_counts()
```

```python
### in gapminder one row = COUNTRY + YEAR
```

```python
### create boxplots to summarize continuous variables (columns)

fig, ax = plt.subplots(figsize=(12, 6))

gap_df.boxplot(ax=ax)

plt.show()
```

```python
### specify the columns to look at

fig, ax = plt.subplots()

gap_df.boxplot(ax=ax, column=['lifeExp'], showmeans=True)

plt.show()
```

```python
### how does the life expectency change across time?

fig, ax = plt.subplots(figsize=(12, 6))

gap_df.boxplot(ax=ax, column=['lifeExp'], showmeans=True, by='year')

plt.show()
```

```python
### we can also make scatter plots to study the relationships between continuous variables
### integers and floats

fig, ax = plt.subplots(figsize=(12, 6))

gap_df.plot.scatter(x='gdpPercap', y='lifeExp', ax=ax)

plt.show()
```

```python
### use a log-scale for the x-axis

fig, ax = plt.subplots(figsize=(12, 6))

gap_df.plot.scatter(x='gdpPercap', y='lifeExp', ax=ax, logx=True)

plt.show()
```

```python
# color by the year

fig, ax = plt.subplots(figsize=(18, 9))

gap_df.plot.scatter(x='gdpPercap', y='lifeExp', c='year', ax=ax, logx=True)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(18, 9))

gap_df.plot.scatter(x='gdpPercap', y='lifeExp', c='year', ax=ax, logx=True, colormap='viridis')

plt.show()
```

```python
## pandas cannot color by a str (object) or discrete variable

fig, ax = plt.subplots(figsize=(18, 9))

gap_df.plot.scatter(x='gdpPercap', y='lifeExp', c='continent', ax=ax, logx=True)

plt.show()
```

## Merging and Joining

```python
### make a DataFrame for countries that I have been to

my_travel_df = pd.DataFrame({'country': ['United States', 'Canada', 'United Kingdom', 'Czech Republic', 'My backyard'],
                            'status': ['live there', 'visited', 'visited', 'visited', 'live there']})

my_travel_df
```

```python
### terminology from SQL - pd.merge(<LEFT DATA>, <RIGHT DATA>, on=<variable to merge on>) -> INNER JOIN
```

```python
my_merged_df = pd.merge( gap_df, my_travel_df, on='country' )
```

```python
my_merged_df.shape
```

```python
my_merged_df['country'].value_counts()
```

```python
### by default pd.merge() performs an INNER JOIN
```

```python
my_merged_df.describe(include=[object])
```

```python
my_merged_df['status'].value_counts()
```

```python
### different type of joint -> LEFT JOIN -> change HOW -> specify the how= argument in pd.merge()

my_left_join_result = pd.merge(gap_df, my_travel_df, on='country', how='left')
```

```python
my_left_join_result.shape
```

```python
my_left_join_result.info()
```

```python
my_left_join_result.head(12)
```

```python
my_left_join_result.loc[ my_left_join_result['country'].str.contains('United') ]
```
