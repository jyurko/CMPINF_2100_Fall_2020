## Visualize relationships between continuous variables

### Using Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
### use the penguins data sets

penguins = sns.load_dataset('penguins')
```

```python
penguins.info()
```

```python
penguins.describe()
```

How does the bill length to relate to the bill depth? Create a scatter plot with `seaborn`, using the FIGURE level function `sns.relplot()`.

```python
sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')

plt.show()
```

```python
sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='scatter')

plt.show()
```

Color or set the `hue` of the marker based on `body_mass_g`.

```python
sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='body_mass_g', 
            kind='scatter')

plt.show()
```

```python
penguins.body_mass_g.nunique()
```

Change the color palette to make it easier to see differences in sizes of penguins. We do this with the `palette` argument and the `sns.color_palette()` function. Please see the documentation for more color choices.

http://seaborn.pydata.org/tutorial/color_palettes.html

```python
sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='body_mass_g',
           kind='scatter',
           palette=sns.color_palette('viridis', as_cmap=True))

plt.show()
```

```python
sns.set_style('darkgrid')

sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='body_mass_g',
           kind='scatter',
           palette=sns.color_palette('viridis', as_cmap=True))

plt.show()
```

```python
sns.set_style('whitegrid')

sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='body_mass_g',
           kind='scatter',
           palette=sns.color_palette('viridis', as_cmap=True))

plt.show()
```

Use a diverging scale.

```python
sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='body_mass_g',
           kind='scatter',
           palette=sns.color_palette('coolwarm', as_cmap=True))

plt.show()
```

Use FACETS or subplots to break up or divide data into categories.

```python
penguins.species.value_counts()
```

```python
sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='body_mass_g', 
           kind='scatter',
           col='species',
           palette=sns.color_palette('viridis', as_cmap=True))

plt.show()
```

Rather than coloring by a continuous variable, we could have colored by the discrete or categorical variable to reveal that relationship.

```python
sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species',
           kind='scatter')

plt.show()
```

```python
sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species',
           kind='scatter',
           palette=sns.color_palette('muted', penguins.species.nunique()))

plt.show()
```

Use a joint distribution or joint histogram to visualize the relationships.

```python
sns.displot(data=penguins, x='bill_length_mm', y='bill_depth_mm', cbar=True)

plt.show()
```

We can use a smoothed joint distribution estimate with a kernel density estimate.

```python
sns.displot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='kde')

plt.show()
```

```python
sns.displot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
           hue='species',
           kind='kde')

plt.show()
```
