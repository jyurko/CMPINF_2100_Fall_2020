## Advanced visualizations of continuous relationships

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
sns.displot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='kde')

plt.show()
```

```python
sns.set_style('whitegrid')
```

```python
sns.displot(data=penguins, x='bill_length_mm', y='bill_depth_mm', 
           hue='species',
           kind='kde')

plt.show()
```

### Joint plot - to show bivariate and univariate behavior

```python
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='kde')

plt.show()
```

```python
sns.kdeplot(data=penguins, x='bill_length_mm')

plt.show()
```

```python
sns.kdeplot(data=penguins, y='bill_depth_mm')

plt.show()
```

```python
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
             hue='species', 
             kind='kde',
             height=8)

plt.show()
```

For greater control we can use the `sns.JointGrid()` to specify different types of plots to make. For example, boxplots to summarize the marginal distributions and a 2D histogram to represent the joint behavior.

```python
g = sns.JointGrid(data=penguins, x='bill_length_mm', y='bill_depth_mm')
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)

plt.show()
```

```python
g = sns.JointGrid(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)

plt.show()
```

```python
penguins.describe()
```

How can we view the relationship between all 4 variables at once?

```python
sns.pairplot(penguins)
```

```python
sns.pairplot(penguins, hue='species')

plt.show()
```

Use `sns.PairGrid()` for greater control of the graphics.

```python
g = sns.PairGrid(penguins)

g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)

plt.show()
```

```python
g = sns.PairGrid(penguins, hue='species')

g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)

plt.show()
```

Sometimes we are just interested in the correlation between each pair of variables. So we need to calculate the correlation coefficient between each pair of variables. We can do that with the `.corr()` method associated with a Pandas DataFrame.

```python
penguin_corr = penguins.corr()
```

```python
type(penguin_corr)
```

```python
penguin_corr.shape
```

```python
penguin_corr
```

We can visualize the correlation matrix as a heatmap with `sns.heatmap()`.

```python
sns.heatmap(penguin_corr, square=True, vmin=-1, vmax=1)

plt.show()
```

Without setting the color bounds...

```python
sns.heatmap(penguin_corr, square=True)

plt.show()
```

It's better to use a DIVERGING scale for correlations.

```python
sns.heatmap(penguin_corr, square=True, vmin=-1, vmax=1,
           cmap=sns.diverging_palette(230, 20, as_cmap=True))

plt.show()
```

```python
sns.heatmap(penguin_corr, square=True, vmin=-1, vmax=1,
            annot=True,
           cmap=sns.diverging_palette(230, 20, as_cmap=True))

plt.show()
```
