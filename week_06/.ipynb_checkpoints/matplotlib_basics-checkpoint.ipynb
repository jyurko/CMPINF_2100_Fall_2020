## Introduction to the Matplotlib figure object

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Two different syntaxes

### MATLAB Style

#### Step 1: initiate the figure object

#### Step 2: specifying a subplot that contains the "data"

#### Step 3: plot() the data

#### Step 4: show the plot

```python
plt.figure() # creating the figure

# create the subplot

plt.subplot(2, 1, 1) # (rows, column, panel "number")
plt.plot(np.sin(np.linspace(0, 10, 100)))

### create a second subplot

plt.subplot(2, 1, 2)
plt.plot(np.cos(np.linspace(0, 10, 100)))

plt.show()
```

### Object Oriented style

#### Step 1: initiate the figure and axes objects

```python
fig, ax = plt.subplots(2)
```

```python
%whos
```

```python
print( len(ax) )
```

```python
### plot a sin wave in the same style we did previously and assign to the
### top panel in the figure
ax[0].plot(np.sin(np.linspace(0, 10, 100)))
### plot a cos wave in the bottom panel
ax[1].plot(np.cos(np.linspace(0, 10, 100)))
```

```python
fig
```

Add axes to the `fig` figure object.

```python
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4]) ### [left, bottom, width, height]

ax2 = fig.add_axes([0.65, 0.65, 0.2, 0.2]) ### [left, bottom, width, height]

ax1.plot(np.sin(np.linspace(0, 10, 100)))

ax2.plot(np.cos(np.linspace(0, 10, 100)))

plt.show()
```

## Grid subplots

```python
fig = plt.figure()

for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)))
```

```python
%matplotlib inline
```

```python
fig = plt.figure()

for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)))
```

```python
fig = plt.figure()

fig.subplots_adjust(hspace=0.4, wspace=0.4)

for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)),
            fontsize=18, ha='center')
```

```python
fig = plt.figure()

plt.subplot(2, 3, 1)


plt.subplot(2, 3, 6)
```

Use the `plt.subplots()` to create the grid all at once.

```python
fig, ax = plt.subplots(2, 3, sharex=True, sharey=True)
```

```python
ax.shape
```

```python
for i in range(2):
    for j in range(3):
        ax[i, j].text(0.5, 0.5, str((i,j)),
                     fontsize=18, ha='center')
```
