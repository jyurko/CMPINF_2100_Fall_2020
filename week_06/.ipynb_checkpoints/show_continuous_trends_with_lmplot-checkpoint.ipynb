## Study trends between continuous variables

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
sns.set_style('whitegrid')
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.info()
```

Visualize the relationship between `price` and `carat`, with the FIGURE level plot `sns.relplot()`.

```python
sns.relplot(data=diamonds, x='carat', y='price')

plt.show()
```

Use transparancy to study where the data are clustered. Transparency is controlled by `alpha`.

```python
sns.relplot(data=diamonds, x='carat', y='price', alpha=0.01, kind='scatter')

plt.show()
```

Use a log-transformation on the scale.

```python
ax = sns.relplot(data=diamonds, x='carat', y='price', alpha=0.01, kind='scatter')

ax.set(xscale='log', yscale='log')

plt.show()
```

Use the marker `hue` to se if the diamond `color` impacts the `price`.

```python
ax = sns.relplot(data=diamonds, x='carat', y='price', hue='color', alpha=0.1, kind='scatter',
                height=9, aspect=1.2)

ax.set(xscale='log', yscale='log')

plt.show()
```

Use the FACET or subplot to study the `clarity`.

```python
ax = sns.relplot(data=diamonds, x='carat', y='price', hue='color', 
                 col='clarity', col_wrap=2,
                 alpha=0.1, kind='scatter',
                height=9, aspect=1.2)

ax.set(xscale='log', yscale='log')

plt.show()
```

How many observations are there for each combination of `cut`, `color`, and `clarity`.

```python
diamonds.groupby(['cut', 'color', 'clarity']).size().reset_index(name='num_rows')
```

```python
sns.displot(data=diamonds, x='color', y='clarity', col='cut', discrete=(True,True), cbar=True)

plt.show()
```

## Use a trend line to visualize a trend

Use `sns.regplot()` or `sns.lmplot()`.

Calculate the log2 transformations of `price` and `carat`.

```python
diamonds['log2_price'] = np.log2(diamonds['price'])
diamonds['log2_carat'] = np.log2(diamonds['carat'])
```

```python
sns.lmplot(data=diamonds, x='log2_carat', y='log2_price')

plt.show()
```

```python
sns.lmplot(data=diamonds, x='log2_carat', y='log2_price',
          scatter_kws={'alpha':0.01})

plt.show()
```

```python
sns.lmplot(data=diamonds, x='log2_carat', y='log2_price',
          hue='color',
          scatter_kws={'alpha':0.001},
          height=8)

plt.show()
```

```python
sns.lmplot(data=diamonds, x='carat', y='price',
          hue='color',
          scatter_kws={'alpha':0.1},
          height=8)

plt.show()
```

Use a non-linear smoother.

```python
sns.lmplot(data=diamonds, x='carat', y='price',
          hue='color',
          scatter_kws={'alpha':0.1},
           lowess=True,
          height=8)

plt.show()
```
