```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
wine_url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

wine_names = ['Cultivar', 'Alcohol', 'Malic_acid', 'Ash', 
              'Alcalinity_of_ash', 'Magnesium', 'Total_phenols',
              'Flavanoids', 'Nonflavanoid_phenols', 'Proanthocyanin', 'Color_intensity',
              'Hue', 'OD280_OD315', 'Proline']
```

```python
wine_data = pd.read_csv(wine_url, names=wine_names)
```

```python
wine_data.info()
```

```python
wine_data['Cultivar'] = wine_data.Cultivar.astype('category')
```

```python
sns.pairplot(wine_data, hue='Cultivar')

plt.show()
```

## PCA

```python
wine_features = wine_data.select_dtypes('number').copy()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = StandardScaler().fit_transform(wine_features)
```

```python
from sklearn.decomposition import PCA
```

```python
pca_wine = PCA(n_components=2).fit_transform(X)
```

```python
print( type(pca_wine) )
```

```python
print( pca_wine.shape )
```

```python
pca_df = pd.DataFrame(pca_wine, columns=['pc1', 'pc2'])
```

```python
sns.relplot(data = pca_df, x='pc1', y='pc2')

plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
clusters_2 = KMeans(n_clusters=2, random_state=121, n_init=25, max_iter=500).fit_predict(X)
```

```python
pca_df['k2'] = pd.Series(clusters_2, index=pca_df.index)
```

```python
sns.relplot(data = pca_df, x='pc1', y='pc2', hue='k2')

plt.show()
```
